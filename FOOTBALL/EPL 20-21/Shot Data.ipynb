```python
results = tableData.drop(['id', 'xG', 'player_id','date'], axis = 1)
```

```python
# results = results.rename(columns={'h_team':'Home','a_team':'Away','h_goals':'Goals(H)','a_goals':'Goals(A)'}, inplace=True) 
```

```python
results.tail()
```

```python
score = results['h_goals']
score_opp = results['a_goals']
```

```python
points_map = {
    'W': 3,
    'D': 1,
    'L': 0
}

def get_result(score, score_opp):
    if score == score_opp:
        return 'D'
    elif score > score_opp:
        return 'W'
    else:
        return 'L'
```

```python
# full time goals
results['Goals'] = np.where(results['h_team'] != results['a_team'],
                                 results['h_goals'],
                                 results['a_goals'])
results['Goals_Opp'] = np.where(results['a_team'] != results['h_team'],
                                     results['a_goals'],
                                     results['h_goals'])
results['Result'] = np.vectorize(get_result)(results['Goals'], results['Goals_Opp'])
results['Points'] = results['Result'].map(points_map)
```

```python
team_results = (results
.sort_values(by=['Date', 'Time']))
```

```python
team_results.head()
```

```python
team_results = team_results.groupby(['match_id', 'Date', 'Time', 'h_team', 'a_team'])['Goals','Goals_Opp','Points'].mean()
```

## Calculating Home Results League Table (Home Standings Table)
